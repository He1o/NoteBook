### 8. <span id = "8">字符串转换整数<span>

字符串转为整数，忽略前置空格，忽略后置字符串。

有限状态机（deterministic finite automaton, DFA）

```python
class automaton():
    def __init__(self):
        self.state = 'start'
        self.sign = 1
        self.ans = 0
        self.table = {
            # space sign digit alpha
            'start': ['start', 'sign', 'digit', 'end'],
            'sign': ['end', 'sign', 'digit', 'end'],
            'digit': ['end', 'end', 'digit', 'end'],
            'end': ['end', 'end', 'end', 'end']
        }
    
    def get_index(self, ch):
        if ch == ' ':
            return 0
        elif ch in ['+', '-']:
            return 1
        elif ch.isdigit():
            return 2
        else:
            return 3

    def get(self, ch):
        self.state = self.table[self.state][self.get_index((ch))]
        if self.state == 'sign':
            if ch == '-':
                self.sign = -1
        elif self.state == 'digit':
            self.ans = self.ans * 10 + int(ch)
            
        
class Solution:
    def myAtoi(self, s: str) -> int:
        auto = automaton()
        for ch in s:
            auto.get(ch)

        ans = auto.ans * auto.sign
        ans = min(ans, 2**31 - 1)
        ans = max(ans, -2**31)

        return ans
```

```python
class Automation{
    string state = "start";
    unordered_map<string, vector<string>> table = {
        {"start", {"start", "signed", "in_number", "end"}},
        {"signed", {"end", "end", "in_number", "end"}},
        {"in_number", {"end", "end", "in_number", "end"}},
        {"end", {"end", "end", "end", "end"}}
    };
    int get_col(char c){
        if (c == ' '){
            return 0;
        }else if (c == '+' or c == '-'){
            return 1;
        }else if (isdigit(c)){
            return 2;
        }else{
            return 3;
        }
    }


public:
    long long res = 0;
    int sign = 1;

    void get(char c){
        state = table[state][get_col(c)];
        if (state == "signed"){
            sign = c == '+' ? 1 : -1;
        }else if (state == "in_number"){
            res = res * 10 + (c - '0');
            res = sign == 1 ? min(res, (long long)INT_MAX) : min(res, -(long long)INT_MIN);
        }
    }
};

class Solution {
public:
    int myAtoi(string s) {
        Automation autom;

        for (char c : s)
            autom.get(c);
        
        return autom.sign * autom.res;

    }
};
```

### 28. <span id = "28">实现 strStr()<span>

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if not needle:
            return 0

        m, n = len(haystack), len(needle)
        i, j = 0, 1
        Pmt = [0]
        while j < n:
            if needle[i] == needle[j]:
                i += 1
                j += 1
                Pmt.append(i)
            elif i == 0:
                j += 1
                Pmt.append(0)
            else:
                i = Pmt[i - 1]

        i = j = 0
        while j < m:
            if needle[i] == haystack[j]:
                i += 1
                j += 1
                if i == n:
                    return j - i
            elif i == 0:
                j += 1
            else:
                i = Pmt[i - 1]
        return -1

    def strStr(self, haystack, needle):
        if not needle:
            return 0
        
        i, j = 0, len(needle)
        while j <= len(haystack):
            if haystack[i:j] == needle:
                return i
            i += 1
            j += 1
        return -1

'''
KMP算法分为两部分

第一部分求pmt部分匹配表
PMT中的值是字符串的前缀集合与后缀集合的交集中最长元素的长度
 j=1
aabaaf
 aabaaf
 i=0
01

 j=1
aabaaf
 aabaaf
  i=1
01


  j=2
aabaaf
  aabaaf
  i=0
010


   j=3
aabaaf
   aabaaf
   i=0
0101
   ↑
   i+1

    j=4
aabaaf
   aabaaf
    i=1
01012
    ↑
    i+1


     j=5
aabaaf
   aabaaf
     i=2
01012

     j=5
aabaaf
    aabaaf
     i=pmt(i-1)=1
01012

aabaaf
     aabaaf
     i=pmt(i-1)=0
010120

有点类似动态规划的思想
我开始想的是相同的话ij都后移一位，这点没有问题
如果不同则该点pmt为0，j后移一位，i=0，这点就有问题了
尽管当前ij的值不同不代表j与i前面的每一位都不同，最起码j还应该i=0进行比较
尽管这样还是会漏掉j前面已经有匹配的元素
那i应该返回到哪一位？
可以假设i前面还有i-1位元素，可以确定的是这些元素与j前面同数量的元素是匹配的
而这些元素的最大相同前后缀就是pmt[i-1]
因此让i=pmt[i-1]就是让i回退到前面i-1个元素的最大相同前后缀
当i=0时与j元素还不匹配，则j后移一位

有了目标pmt表
j在原字符串上移动
同理
注意j和i都是从0开始
且判断i==n说明完全匹配返回j-n
'''
```
